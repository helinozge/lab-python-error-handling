def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))

````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(products):
    inventory = {}
    for product in products:
        while True:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    print("Quantity cannot be negative. Try again.")
                else:
                    inventory[product] = quantity
                    break
            except ValueError:
                print("Invalid input. Please enter a whole number.")
    return inventory

def get_customer_orders(inventory):
    while True:
        try:
            num_orders = int(input("Enter the number of customer orders: "))
            if num_orders < 0:
                print("Number of orders cannot be negative.")
                continue
            break
        except ValueError:
            print("Invalid input. Please enter a valid number.")
    
    customer_orders = set()
    for _ in range(num_orders):
        while True:
            product = input("Enter a product the customer wants to order: ").strip().lower()
            if product not in inventory:
                print(f"Invalid product. Available products: {list(inventory.keys())}")
            elif inventory[product] <= 0:
                print(f"Sorry, {product} is out of stock.")
            else:
                customer_orders.add(product)
                break
    return customer_orders

def calculate_total_price(customer_orders):
    prices = {}
    for product in customer_orders:
        while True:
            try:
                price = float(input(f"Enter the price of {product}: "))
                if price < 0:
                    print("Price cannot be negative.")
                else:
                    prices[product] = price
                    break
            except ValueError:
                print("Invalid input. Please enter a numeric price.")
    return sum(prices.values())

def update_inventory(customer_orders, inventory):
    for product in customer_orders:
        if product in inventory and inventory[product] > 0:
            inventory[product] -= 1
    return {k: v for k, v in inventory.items() if v > 0}

def calculate_order_statistics(customer_orders, products):
    total = len(customer_orders)
    percent = (total / len(products)) * 100
    return total, percent

def print_order_statistics(stats):
    total, percent = stats
    print("\nOrder Statistics:")
    print(f"Total Products Ordered: {total}")
    print(f"Percentage of Products Ordered: {percent:.2f}%")

def print_updated_inventory(inventory):
    print("\nUpdated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")

# ----- Main Program -----
inventory = initialize_inventory(products)
customer_orders = get_customer_orders(inventory)
print_order_statistics(calculate_order_statistics(customer_orders, products))
inventory = update_inventory(customer_orders, inventory)
print_updated_inventory(inventory)
total_price = calculate_total_price(customer_orders)
print(f"\nTotal Price: {total_price:.2f}")
````
